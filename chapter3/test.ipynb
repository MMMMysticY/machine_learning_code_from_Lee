```python
import numpy as np
import torch
import torch.nn as nn
```

```python
m = nn.Conv2d(3,64,3,1,1)
input = torch.randn(20,3,128,128)
input = input + 10
output = m(input)
output.shape
```

```python
maxpool = nn.MaxPool2d(2,2)
output_ = maxpool(output)
output_.shape
```

```python
linear = nn.Linear(64*64*64,11)
output_ = output_.flatten(1)
output__ = linear(output_)
output__
```
